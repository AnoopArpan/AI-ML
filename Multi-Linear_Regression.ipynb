```python
x=data.loc[0:,["YearsExperience","Age"]]
y=data.loc[:,"Salary"]
x
```

```python
### Split Data in train test and variables
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr_model=LinearRegression()
```

```python
mlr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
x_pred=lr_model.predict(x_train)
```

```python
y_pred,y_test
```

```python
x_pred,y_train
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
print("TRAINING ACCURACY:",r2_score(y_train,x_pred)*100)
print("TESTING ACCURACY:",r2_score(y_test,y_pred)*100)
```
