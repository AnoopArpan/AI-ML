```python
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,'ro')
plt.plot(x,y,color="green")
plt.title("Graph")
plt.xlabel("Index no.")
plt.ylabel("Dataset")
plt.show()
```

```python
###bar graph
student=["A","B","C","D"]
marks=[24,43,49,32]
plt.bar(student,marks,0.4,color="red")
plt.title("Python result")
plt.xlabel("Students name")
plt.ylabel("Marks")
plt.show()
```

```python
student=["A","B","C","D"]
marks=[24,43,49,32]
plt.pie(marks,labels=student,colors=["blue","red","orange","brown"])
plt.title("Python result")
plt.show()
```

```python
student=["A","B","C","D"]
student1=["E","F","G","H"]
marks_python=[24,43,49,32]
marks_maths=[22,43,33,28]
plt.bar(student,marks_python,0.4,color="red")
plt.bar(student1,marks_maths,0.4,color="orange")
plt.title("Python result")
plt.xlabel("Students name")
plt.ylabel("Marks")
plt.show()
```

```python
student=["A","B","C","D"]
marks=[24,43,49,32]
exp=[0.2,0,0.2,0]
plt.pie(marks,labels=student,colors=["blue","red","orange","brown"],explode=exp,shadow=True)
plt.title("Python result")
plt.show()
```
