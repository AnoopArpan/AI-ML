### Real-Time Sign Detection with Webcam

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Define region of interest (ROI) for hand gesture
    x1, y1, x2, y2 = 100, 100, 300, 300
    roi = frame[y1:y2, x1:x2]

    # Draw ROI rectangle
    cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Predict sign language gesture
    processed = preprocess_image(roi)
    prediction = model.predict(processed)[0]
    predicted_class = classes[np.argmax(prediction)]
    confidence = np.max(prediction)

    # Display result
    text = f"Letter: {predicted_class} ({confidence:.2f})"
    cv2.putText(frame, text, (50, 400), cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)

    # Show the frame
    cv2.imshow("Sign Language Converter", frame)

    # Break on 'q' key
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
