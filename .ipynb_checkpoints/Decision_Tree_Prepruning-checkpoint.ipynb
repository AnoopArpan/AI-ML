```python
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=load_iris()
data1=sns.load_dataset("iris")
data
```

```python
data1
```

```python
x=data["data"]
y=data["target"]
```

```python
x=data1.iloc[:,:-1]
y=data1.iloc[:,-1]
```

```python
y
```

```python
x
```

```python
x.ndim
```

```python
le=LabelEncoder()
```

```python
le.fit_transform(data1["species"])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=23)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
para={
        'criterion':['gini','entropy','log_loss'],
        'splitter':['best','random'],
         'max_depth':[1,2,3,4,5,6]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gd_model=GridSearchCV(model,param_grid=para,cv=3,scoring="accuracy")
```

```python
gd_model.fit(x_train,y_train)
```

```python
gd_model.best_params_
```

```python
from sklearn import tree
```

```python
y_pred=gd_model.predict(x_test)
x_pred=gd_model.predict(x_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print("testing",accuracy_score(y_test,y_pred))
print("Training",accuracy_score(y_train,x_pred))
```
