```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=42)   #or test size=0.30
```

```python
### linear regression algorithm
-y=mx+c
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred= lr_model.predict(x_test)
```

```python
y_pred,y_test
```

```python
x_pred=lr_model.predict(x_train)
```

```python
plt.scatter(x_train,y_train,color="green")
plt.plot(x_train,x_pred,color="red")
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred,color="red")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

```python
print("TRAINING ACCURACY:",r2_score(y_train,x_pred)*100)
print("TRAINING ACCURACY:",r2_score(y_test,y_pred)*100)
```
