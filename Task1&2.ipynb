### TASK 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Titanic-Dataset.csv")
```

```python
data
```

```python
data["Embarked"].unique()
```

```python
data["Sex"]=data["Sex"].map({'male':0,'female':1})
```

```python
data["Embarked"]=data["Embarked"].map({'S':0,'C':1,'Q':2})
```

```python
x=data[["PassengerId"]]
y=data["Fare"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
x_pred=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
r2_score(y_test,y_pred)*100
```

### Multi Linear regression

```python
x=data[["PassengerId","Survived","Pclass","SibSp","Parch","Sex"]]
y=data["Fare"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70,random_state=42)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
x_pred=lr.predict(x_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
r2_score(y_test,y_pred)*100
```

### Kmean

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=1000,n_features=2,centers=1)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
from sklearn.cluster import KMeans
KMean=KMeans(n_clusters=3)
```

```python
y_pred=KMean.fit(x_train)
```

```python
y_pred
```

```python
x_pred=KMean.predict(x_train)
x_pred
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=x_pred)
plt.show()
```

```python
x_test_pred=KMean.predict(x_test)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c=x_test_pred)
plt.show()
```
